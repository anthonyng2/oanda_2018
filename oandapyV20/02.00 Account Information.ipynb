## Account Details

```python
import pandas as pd
import oandapyV20
from oandapyV20 import API
import oandapyV20.endpoints.accounts as accounts
import configparser
```

```python
accountID = "101-003-5120019-001"
access_token = "57a93815fcfbfbe19baea799bc144c15-f18df92fa9762978909559704fdfff60"

api = API(access_token=access_token)
```

```python
client = oandapyV20.API(access_token=access_token)
r = accounts.AccountDetails(accountID)
```

```python
client.request(r)
```

```python
print(r.response)
```

```python
pd.Series(r.response['account'])
```

## Account List

```python
r = accounts.AccountList()
```

```python
client.request(r)
```

```python
print(r.response)
```

## Account Summary

```python
r = accounts.AccountSummary(accountID)
```

```python
client.request(r)
```

```python
print(r.response)
```

```python
pd.Series(r.response['account'])
```

## Account Instruments

```python
r = accounts.AccountInstruments(accountID=accountID, params = "EUR_USD")
```

```python
client.request(r)
```

```python
pd.DataFrame(r.response['instruments'])
```
